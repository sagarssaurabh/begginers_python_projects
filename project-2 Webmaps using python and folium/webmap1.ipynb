```python
names = list(df1.NAME)
location = list(df1.LOCATION)
elevation = list(df1.ELEV)
```

```python
def color_produce(elevation):
    if elevation >= 0 and elevation <2000:
        return "green"
    elif elevation >= 2000 and elevation <3000:
        return "orange"
    elif elevation >= 3000:
        return " red"
```

```python
#--------------two ways to add information in marker--------
# for info in list(zip(zip(list1,list2),zip(names,location,elevation))): #so in info we have gathered all information of a coordinates along with their location and thier elevaton
#     map1.add_child(folium.Marker(location = info[0],popup = info[1],icon = folium.Icon(color = "green")))

#---------along with adding marker , here we are adding different colour tags depending upon elevation------------
# for lat,lon,nm,add,el in zip(list1,list2,names,location,elevation): 
#     map1.add_child(folium.Marker(location = [lat,lon],popup = str(nm)+ ", " + str(add)+ ", " + str(el) + " m",icon = folium.Icon(color = color_produce(el))))

#-------------------we are converting tags to circle and add some opacity to them-------------------
for lat,lon,nm,add,el in zip(list1,list2,names,location,elevation): 
    map1.add_child(folium.CircleMarker(location = [lat,lon],radius=6,popup = str(nm)+ ", " + str(add)+ ", " + str(el) + " m",fill_color = color_produce(el),color = "grey",fill_opacity = 0.7))

    
```

```python
#-----We will use GIS - Geographical Information systems--------
# ----so there are two layers in this map, one is land or sea region shown li map
#.  second layer is the layers on which markers are made

# ---now for each of the country we will have specific colour depending upon the population of that country
# ---now area is represented by polygon so to represent polygon using Folium we will use GeoJson methon
map1.add_child(folium.GeoJson(data=(open('world.json', 'r'))))
```

```python
map1.save("volcanoes1.html")
```
